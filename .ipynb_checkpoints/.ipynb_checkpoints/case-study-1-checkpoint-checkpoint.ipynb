***Task 1: Choose your favourite dataset***

1. Choice of Study: Virtual communication curbs creative idea generation

2. Summary:

    This study explores the impact of the remote work on creativity. Previously, we relied on emails and phone calls as ways to exchange information. Many theorized this communication technology to be limiting the extent of information conveyed to others and lacking synchronicity. Recent advances in network quality and display resolution has lead to more synchronous communication, with face-to-face capturing non-verbal and aural information cues which normally wouldn't with previous communication technology.


    This study addresses the question of whether recent virtual technology could replace workplace in-person idea generation. Conclusively, the study found virtual communication -- specifically videoconferencing-- inhibits creative idea generation. The in-person groups had better idea generation performance (number of ideas, quality of ideas) than virtual groups. To find possible causes as to why, they recorded the amount of people's eyegaze, recall measures and latent semantic analysis during the exercises and found these measures positively correlated with creative idea generation. 

    So, even though there are advantages and lesser limitations in virtual collaboration than prior, we can't truly say that it can fully replace in-person collaboration when it comes to generating creative ideas. These findings could inform the design of remote work policies considering that 75% of US employees expressed their preference in working at home at least one day a week. Although the results demonstrated an advantage to working in-person, the study still considers the immediate economic and practical advantages virtual collaboration offers such as having to deal with different time zones and geographic region, and the cost of commute, and business travel expenses. The study only documents the cognitive cost of virtual interaction and suggests a more comprehensive analysis in the future including context-specific and industry costs.

**Task 3:** Data Visualization

8. Two numerical variables were selected: `partner_time` and `task_time`.

To explore the relationship between these two numerical variables, use a scatter plot. We are also going to differentiate by whether the individual was assigned to be a writer or not by adding `color = writer` in **aes()** function.

```r
virt_plot <- virt_comm_data %>%
    ggplot(aes(x = partner_time, y = task_time, color = writer)) + 
    geom_point() +  
    geom_smooth(method = "lm", se = FALSE) +
    labs(title = "Relationship between time spent gazing at task and at partner", 
         x = "Partner Time (seconds)", 
         y = "Task Time (seconds)",
         color = "Writer?") + 
    scale_colour_discrete(labels = c("FALSE" = "non-writer", "TRUE" = "writer")) + 
    theme_minimal()

virt_plot
```

The scatterplot reveals a strong negative relationship indicating that the more time spent gazing at the partner (partner_time), the less time spent gazing at the task computer (task_time) regardless whether you're the writer or not. The parallel lines with negative slopes indicate shared slopes i.e. rate at which partner time increases as task time decreases is the same for both writers and non writers. However, they have different baselines, with the writers showing higher y intercept than non writers.

17. Considering your dataset and research question, create a new feature derived from the existing variables in your dataset that could help you answer your question.

We could construct a new variable that calculates the total time per observation using **mutate()** to create a new column `total_time`. Additionally, we can create three new columns `task_prop`,`partner_prop`,`room_prop` to get the proportion of each activity within total time allocated by dividing the numeric variable by `total_time`. This allows for us to see the difference in proportions of time spent in each activity across different conditions and compare how participants allocate time differently. See code below.

```r
virt_ave_time <- virt_comm_data %>%
    select(cond, task_time, partner_time,room_time) %>%
    mutate(total_time = task_time + partner_time + room_time,
           task_prop = task_time/total_time,
           partner_prop = partner_time/total_time,
           room_prop = room_time/total_time) 

virt_box2 <- virt_ave_time %>%
    ggplot(aes(x = cond, y = room_prop)) +
    geom_boxplot() +
    labs(title = "Total time on task", 
         x = "Experimental Condition", 
         y = "Total Time (seconds)") + 
    theme_minimal()

virt_box2
    

virt_ave_time
```

20. Was this an observational study or controlled experiment study? This was a controlled study because the researchers assigned participants to certain conditions (face-to-face, videoconferencing). They manipulated a variable and controlled the setting to see an effect.

21. Did researchers control for any confounding variable in this study? If yes explain how. If no, can you think of any potential confounding variable that you can adjust for? They found that those pairs who worked on the task face-to-face looked around the room and at their partner more, and produce more creative ideas. So, they inferred that working face-to-face produces more creative ideas. Some confounding variables could be familiarity with the tasks given, or expertise of the participants. For example, what if those who tend to look around more or interact with their partners are the experts of the task? What if it's not the amount of gazing time that causes higher level of creativity but the person's expertise? Researchers controlled for this variable by doing a field experiment on experts such as engineers as opposed to just novice students and they found similar results. 

22. They used convenience sampling (Brucks & Levav, 2022, p. 109 *Laboratory experiment* section). This makes it easy for the authors to have easy access to sampling data because they are in close connections to students as academics and these students are most likely to have prior experiences in working in pairs or team environment to complete a task. They're also the demographic who's adept at using computers and audio-visual technology.

23. Non-students and teens may be underrepresented and thus create sampling bias. Many non-students and teens also use audio-visual technology to communicate and innovate ideas and they weren't taken accounted for in the sample. Also, they have chosen engineers to represent the "experts" to communicate that they share similar results as the "novice" or students. This could be different for other groups people in different industries i.e. finance, arts.
